```python
# histograma e distribuição cumulativa
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('unequal.png',0)

hist,bins = np.histogram(img.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.imshow(img, cmap = 'gray', clim=(0, 255))
plt.show()

plt.plot(cdf_normalized, color = 'b')
plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
#equalização do histograma
equ = cv2.equalizeHist(img)

plt.subplot(121),plt.imshow(img, cmap = 'gray', clim=(0, 255))
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(equ, cmap = 'gray', clim=(0, 255))
plt.title('Image after equalization'), plt.xticks([]), plt.yticks([])
plt.show()
```
